# **Linear Regression - Line Graph** **Plotting**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0,40,1000)
```

```python
ds_r = 2.5*t
ds_s = 3*(t-5)
```

```python
fig,ax = plt.subplots()
plt.title("Rober Caught Graph")
plt.xlabel("Time (in min)")
plt.ylabel("Distance (in km)")
ax.set_xlim([0,40])
ax.set_ylim([0,100])
ax.plot(t,ds_r,c="red")
ax.plot(t,ds_s,c='green')
plt.axvline(x=30,color='purple',linestyle="-.")
plt.axhline(y=75,xmin=0,xmax=30,color='purple',linestyle="-.")
```
